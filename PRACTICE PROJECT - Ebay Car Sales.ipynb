```
# Renaming and clarifying some of the column names first
columns = autos.columns
new_columns = []
for c in columns:
  if c == 'yearOfRegistration':
    c = 'registration_year'
    new_columns.append(c)
    print(new_columns)
  elif c == 'monthOfRegistration':
    c = 'registration_month'
    new_columns.append(c)
    print(new_columns)
  elif c == 'notRepairedDamage':
    c = 'unrepaired_damage'
    new_columns.append(c)
    print(new_columns)
  elif c == 'dateCreated':
    c = 'ad_created'
    new_columns.append(c)
    print(new_columns)
  else:
    new_columns.append(c)
```

```
# Addressing camelcase to snakecase issue next
import re
cleaned_columns = []
for c in new_columns:
  c = re.sub('([A-Z]+)', r'_\1',c).lower()
  cleaned_columns.append(c)
```

In the next section we are going to aggregate data based on the various car brands.
I'm going to run through the process in the following steps:
* Identify values that can be of interest
* Create a dictionary to store the values in
* find the means of the data to compare the brands with each other

```
autos['brand'].unique()
```

```
autos['brand'].value_counts()
```

I think it would be quite interesting to be able to aggregate by country as well but that data is not in this dataset, so i'm going to go with every brand that has more than a 5% share of all the ads instead.
I like that method better than just going with the top X amount of ads.

```
# Setting up our new dataset to work and perform analysis with
autos.head()
```

```
# picking all the brands that have more than 5% share of all the ads and put it in a dictionary
brands = autos['brand'].value_counts(normalize=True) > 0.05
```

```
top_brands = {}
for b in brands.index:
  if brands[b] == True:
    top_brands[b] = None
```

```
# We now have a Dictionary with all the top brands to work with.
top_brands
```

Just looking at the brand names i'm noticing that 5 out of the 6 brands are german brands. Since this data was sourced from the german version of ebay, this is not surprising but should be something to be taken into account when evaluating country specific data.

Next step is to collect the average data of each brand. 
I personally would be intreested in 'power_ps', 'odometer_km' and 'price'
There are lots of great insights that could be gained from all those columns, but i'm going to stick to the guided project and simply find the mean price of all those brands.

```
# separate out all the prices for each brand and calculate the mean
for key in top_brands:
  top_brands[key] = []
  top_brands[key].append(autos[autos['brand'] == key]['price_$'].mean())
```

```
# output our dictionary of values in descending
sorted(top_brands.items(),key=lambda x: x[1],reverse=True)
```

Unsurprisingly, the premium brands Audi, Mercedes and BMW have a much larger resale value than the mid range brands Ford and Opel while Volkswagen who by far has the most ads up is right in the middle of the range.
